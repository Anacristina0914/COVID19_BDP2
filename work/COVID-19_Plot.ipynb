```python
import requests
import json 
from matplotlib import pyplot as plt
```

```python
r =requests.get("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-andamento-nazionale.json")
j = json.loads(r.content)
```

```python
fx = "data"
fy ="nuovi_positivi"


x = []
y = []

for i in j:
    x.append(i[fx])
    y.append(i[fy])

plt.title(fy)
plt.plot(x,y)
```
